```python
import pandas as pd
cust_df = pd.read_csv("CustomerData.csv")
cust_df.head()
```

```python
df = cust_df.drop('Address', axis=1)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

X = df.values[:,1:]
X = np.nan_to_num(X)

Clus_dataSet = StandardScaler().fit_transform(X)
Clus_dataSet
```

```python
clusterNum = 3
```

```python
k_Means = KMeans(n_clusters=clusterNum,n_init=12,init='k-means++')
k_Means.fit(X)
labels = k_Means.labels_
print(labels)
```

```python
df["Clus_km"] = labels
df.head(5)
```

```python
df.groupby('Clus_km').mean()
```

```python
area = np.pi * ( X[:, 1])**2  
```

```python
plt.scatter(X[:, 0], X[:, 3], s=area, c=labels.astype(np.float64), alpha=0.5)
plt.xlabel('Age', fontsize=18)
plt.ylabel('Income', fontsize=16)

plt.show()
```
